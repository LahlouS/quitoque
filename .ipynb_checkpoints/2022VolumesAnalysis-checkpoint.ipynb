# 2022 sales volumes analysis

### let's start with the big picture, a line graph of sales per day on the all year

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates


# parsing the file into a panda dataframe
day_dataframe = pd.read_excel('./Cas_pratique_-_Volumes_par_jour.xlsx', sheet_name='Commandes', \
                              parse_dates={'day': ['Jour']}) # loading into date format

# configuring the graph size
plt.figure(figsize=(12, 6))

plt.plot(day_dataframe['day'], day_dataframe['Nombre de commandes'])

plt.xlabel('Date')
plt.ylabel('Volumes')
plt.title('Daily Volumes')

# Specify the desired date range for the x-axis
start_date = day_dataframe['day'].min()
end_date = day_dataframe['day'].max()

plt.xlim(start_date, end_date)

# Set the locator and formatter for the x-axis dates
date_locator = mdates.MonthLocator()
date_formatter = mdates.ConciseDateFormatter(date_locator)

# Apply the locator and formatter to the x-axis
plt.gca().xaxis.set_major_locator(date_locator)
plt.gca().xaxis.set_major_formatter(date_formatter)

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()


```

### now zooming into week volumes summary

```python
import pandas as pd
import glob

# Get a list of SXX files paths
excel_files = glob.glob('./S*.xlsx')

# Create an empty list to store DataFrames
dataframesWeeksum = []

# Read each Excel file into a DataFrame and append to the list
i = 0
while i < 53:
    if i < 10:
        file = f'S0{i}.xlsx'
    else:
        file = f'S{i}.xlsx'
    df = pd.read_excel(file, sheet_name='week_sum')
    dataframesWeeksum.append(df)
    i += 1
    
# Concatenate the DataFrames into a single DataFrame
combined_df = pd.concat(dataframesWeeksum)

# Reset the index of the combined DataFrame
combined_df = combined_df.reset_index(drop=True)

plt.plot(combined_df.index.values, combined_df['total_volume'], marker='o', label='vol_per_week')
plt.plot(combined_df.index.values, combined_df['average_per_day'], marker='o', label='av_per_week')

plt.xlabel('weeks')
plt.ylabel('volume')

plt.title('average and total volume per weeks on 2022')

plt.legend()

plt.show()
```

### finally lets zoom into days sells

```python
import pandas as pd
import glob

# Get a list of SXX files paths
excel_files = glob.glob('./S*.xlsx')

# Create an empty list to store DataFrames
dataframeDaySum = pd.DataFrame({'day_sum': [0, 0, 0, 0, 0, 0, 0], \
                                'day_av': [0, 0, 0, 0, 0, 0, 0]}, \
                               index=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])


# Read each Excel file into a DataFrame and append to the list
for file in excel_files:
    df2 = pd.read_excel(file, sheet_name='volume_per_days')
    for index, row in df2.iterrows():
        dataframeDaySum.loc[row['day'], 'day_sum'] += row['volume']
        dataframeDaySum.loc[row['day'], 'day_av'] = dataframeDaySum.loc[row['day'], 'day_sum'] / 52

print(dataframeDaySum)
        
plt.plot(dataframeDaySum.index, dataframeDaySum['day_sum'], label='total_volume')
plt.plot(dataframeDaySum.index, dataframeDaySum['day_av'], label='av_per_day')

plt.xlabel('days')
plt.ylabel('volumes')

plt.title('average and total volume counts for each days')

plt.legend()

plt.show()
```

1. **Monthly Sales Volume:** The first graph provides insights into the average volume and standard deviation of daily sales throughout 2022. We observe a typical sales volume ranging between 20 and 40 units per day. However, there is a noticeable dip in sales during the summer months, specifically in July and August, where the volume stabilizes within the range of [10, 30]. Additionally, we notice a recurring pattern of sales throughout the week, characterized by regular peaks outside the usual range. At this point, it is challenging to provide more precise analysis without further context.

2. **Weekly Sales Patterns:** The analysis of weekly sales reveals intriguing patterns. We observe a distinct trend where the number of sales experiences a significant decline approximately every 6-7 weeks, followed by a quick recovery and continuation for another 6-7 week cycle. The script below the weekly analysis highlights specific weeks related to holidays, such as winter, spring, Toussaint, and Noel. These holiday periods directly impact the sales volume.

3. **Daily Sales Count:** Finally, we examine the volume of sales for each day of the week. Tuesdays stand out as the day with the highest sales count, while Sundays consistently show a count of 0 sales. However, drawing definitive conclusions solely based on the Sunday observation would be premature. It is essential to consider potential underlying systemic factors that may impact sales and the specific dynamics of the business.
